```python
import random
import numpy as np
import matplotlib.pyplot as plt
```

```python
f = 0.2

```

```python
n = 10
fs = [0, 0.2, 0.5, 0.8, 1]
ps = [3 , 5, 7]
iterations = 10000

test_rankings = np.zeros((5,3,iterations,n))

for i, f in enumerate(fs):
    for j, p in enumerate(ps):
        for iteration in range(iterations):
            protected = np.ones(p)
            non_protected = np.zeros(n - p)
            ranking = np.zeros(n)

            for rank in range(n):
                r = random.random()
                element = None
                if(r < f):
                    element, protected = protected[:1][0], protected[:-1]
                else:
                    element, non_protected = non_protected[:1][0], non_protected[:-1]    
                ranking[rank] = element

                if(len(protected) == 0):
                    ranking[rank+1:] = non_protected
                    break
                elif(len(non_protected) == 0):
                    ranking[rank+1:] = protected
                    break            
            
            test_rankings[i][j][iteration] = ranking
```

```python
fig, axes = plt.subplots(len(ps),len(fs), sharey=True, sharex=True, figsize=(8,5), constrained_layout=True)
for i in range(len(ps)):
    for j in range(len(fs)):
        axes[i][j].bar(range(1, n+1), test_rankings.sum(axis=2)[j][i] / iterations)        
        axes[i][j].set_xticks(range(1, n+1, 2))

for i in range(len(ps)):
    axes[i][0].yaxis.set_label_coords(-0.5, 0.5)
    axes[i][0].set_ylabel('p = {}'.format(ps[i] / n), rotation='horizontal')

for j in range(len(fs)):
    axes[len(ps)-1][j].set_xlabel('f = {}'.format(fs[j]))
        
plt.show()
```
