```python
KL_DIVERGENCE = "rKL"
```

```python
max_rKL = getNormalizer(user_N,pro_N,KL_DIVERGENCE)  
```

```python
cut_point = 10
```

```python
test_ranking = [x for x in range(user_N)]
```

```python
pro_index = [x for x in range(0, user_N, 2)] + [x for x in range(1, user_N + 1, 2)]
```

```python
print(test_ranking)
print()
print(pro_index)
```

```python
results = []
for i in range(1, user_N):
    max_rKL = getNormalizer(user_N, i, KL_DIVERGENCE)  
    results.append(calculateNDFairness(test_ranking, pro_index[:i], cut_point, KL_DIVERGENCE, max_rKL))
```
